```python
movies=df[['genres','id','keywords','title','overview','cast','crew']]
```

```python
import ast
# ast.literal_eval is used to convert data into its origional form
def convert(obj):
    lis=[]
    obj=ast.literal_eval(obj)
    for i in obj:
        lis.append(i['name'])
    return lis
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast
# ast.literal_eval is used to convert data into its origional form
def convertCast(obj):
    lis=[]
    obj=ast.literal_eval(obj)
    i=0
    while(i<3 and i<len(obj)):
        lis.append(obj[i]['name'])
        i+=1
    return lis
```

```python
import ast
def coverterCrew(obj):
    obj=ast.literal_eval(obj)
    ls=[]
    for i in obj:
        if i['job']=='Director':
            ls.append(i['name'])
            break
    return ls
```

```python
movies['cast']=movies['cast'].apply(convertCast)
```

```python
movies['crew']=movies['crew'].apply(coverterCrew)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split('\n'))
```

```python
movies.head()['overview']
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['overview']=movies['overview'].apply(lambda x:[i.split(' ') for i in x][0])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
movies.drop(columns=['genres','keywords','overview','cast','crew'],inplace=True)
```

```python
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies.head()
```

```python
movies['tags']=movies['tags'].apply(lambda x:x.lower())
movies['tags'].head()
```

```python
# nltk is a very popular natural processing library
# [loves,love,loving,loved]=[love,love,love,love]
# [stems,stem]=[stem,stem] # [tree,tree's]=[tree,tree]
import nltk # Natural Language Toolkit
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
# import nltk # natural language toolkit
# from nltk.stem.porter import PorterStemmer
# ps=PorterStemmer()
# ps.stem("loved")
```

```python
movies['tags']=movies['tags'].apply(stem)
print(movies['tags'].head())
```

```python
# WILL COUNT THE FREQUENCY OF THE WORD BY AVOIDNG THE STOP WORD(LIKE ARE,THE,IS ETC)
# IT WILL MAKE THE FREQUENCY TABLE OF THE SENTENCE in the tag 
# similarly it checks the frequency of the world in other movies tags 
# from the frequency tables it will make the vectors 
# and if we want to recomment five movies on the basis of the current movies it will
# give the 5 closses distance vectors
# the basic idea is to reprent the movies in the form of vectors
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#  by default the cv.fit_tranform give us a sparse array so we will convert it into np array
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
a=[i for i in cv.get_feature_names_out()]
a
```

```python
vectors.shape
```

```python
# euclidean distance is not a good measure in high dimension
# so we find cosine distance but here we use cosine similarity because it is between zero and one
# cosine similariy is the inverse of consine distance
# if the cosine distance is less cosine similarity is more
from sklearn.metrics.pairwise import cosine_similarity 
similarity=cosine_similarity(vectors) #to find the distance of each vector from all the other vectors
print(similarity.shape)
print(similarity[0])
```

```python
movies.head()
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    recommended_movies=[]
    for i in movies_list:
        recommended_movies.append(movies.iloc[i[0],1])
        print(movies.iloc[i[0],0])
    return recommended_movies
```

```python
recommend('Team America: World Police')
```
